```python
def annotate_text_with_gazetteers(text):
    # Initialize annotated text
    annotated_words = []

    # Tokenize the text (simple whitespace-based tokenization)
    words = text.split()

    for word in words:
        # Determine the label based on gazetteers
        if word in diseases:
            label = "B-Disease"
        elif word in symptoms:
            label = "B-Symptom"
        elif word in persons:
            label = "B-Person"
        elif word in consumables:
            label = "B-Consumable"
        else:
            label = "O"

        annotated_words.append(f"{word} {label}")

    # Join annotated words into a single string
    return " ".join(annotated_words)
```

```python
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForTokenClassification, TrainingArguments, Trainer

def convert_to_conll_format(input_file_path, output_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as infile, open(output_file_path, 'w', encoding='utf-8') as outfile:
        for line in infile:
            line = line.strip()
            if line and line[0] != '-':  # Skip lines starting with '-'
                tokens_labels = line.split()
                for token_label in tokens_labels:
                    if token_label == 'O':
                        token, label = token_label, 'O'
                    else:
                        token, label = token_label.rsplit(' ', 1)
                    outfile.write(f"{token}\t{label}\n")
                outfile.write("\n")  # Separate sentences with an empty line

# Convert your file to CoNLL format
input_file_path = "/content/hindi_health_dataset/Annotated_Hindi_Health_Data.txt"
output_file_path = "converted_data.conll"
convert_to_conll_format(input_file_path, output_file_path)
```

```python
dataset = '/content/hindi_health_dataset/Annotated_Hindi_Health_Data.txt'

# Load tokenizer and model
model_name = "google/muril-base-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name, num_labels=len(set(label for sublist in dataset['labels'] for label in sublist)))

# Tokenize the dataset
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples['sentence'], truncation=True, is_split_into_words=True)
    labels = []
    for i, label in enumerate(examples['labels']):
        word_ids = tokenized_inputs.word_ids(i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx])
            else:
                label_ids.append(label[word_idx])
            previous_word_idx = word_idx
        labels.append(label_ids)
    tokenized_inputs["labels"] = labels
    return tokenized_inputs

tokenized_datasets = dataset.map(tokenize_and_align_labels, batched=True)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,
    eval_dataset=tokenized_datasets,
    tokenizer=tokenizer,
)

# Fine-tune the model
trainer.train()

# Save the model and tokenizer
model.save_pretrained("fine-tuned-muril-base-cased")
tokenizer.save_pretrained("fine-tuned-muril-base-cased")
```
